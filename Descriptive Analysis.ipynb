### Total population fully, partially vaccinated and received boosters

```python
# Load malaysia vaccination dataset
# This dataset includes all age groups, nationalities, ethinicities 
total_vax = pd.read_csv(folder + 'vaccination/vax_malaysia.csv')
total_vax
```

```python
list(total_vax)
```

```python
# Calculate vaccinated rate. 
total_pvax = total_vax['cumul_partial'].iloc[-1]
perc_pvax = total_pvax/ms_total_pop*100
print(perc_pvax, 'of population are partially vaccinated.')

total_fvax = total_vax['cumul_full'].iloc[-1]
perc_fvax = total_fvax/ms_total_pop*100
print(perc_fvax, 'of population are fully vaccinated.')

total_boost = total_vax['cumul_booster'].iloc[-1]
perc_boost = total_boost/ms_total_pop*100
print(perc_boost, 'of population received booster.')
```

```python
# Plot a graph to visualize

# Set figure size, varaibles to plot
fig, ax = plt.subplots(1,1,figsize=(10,6))
group = ['Partial','Fully','Booster']
percentage = [perc_pvax, perc_fvax, perc_boost]

# Plot horizontal bar chart with the range of x-axis defined
plt.barh(group, percentage, color= 'lightgreen')
plt.xlim([0,100])

# Add x-axis, y-axis, and title
plt.xlabel('% of Population Vaccinated')
plt.ylabel('Status')
plt.title('Percentage of Population Partial, Fully Vaccinated and Received Booster')

# Add a vertical line at 80% (targeted immunization rate)
plt.axvline(x = 80, color = 'r')

# Add value labels to each bar 
for index, value in enumerate(percentage):
    plt.text(value+1, index, str(round(value, 2)), fontweight='bold')

plt.show()

#fig.savefig("perc of pop vaccinated.png")
```

### Total population fully, partially vaccinated and received boosters based on age groups

```python
# Load vaccination among age groups dataset
vax_age = pd.read_csv(folder + 'vaccination/vax_demog_age.csv')
vax_age
```

```python
list(vax_age)
```

```python
# Calculate daily vaccination administrated for vax_age dateset
new_vax_age = vax_age.groupby('date').sum()
new_vax_age
```

```python
age = new_vax_age.sum()
age.sort_index(inplace=True)
age
```

```python
# Calculate total vaccinated population for age groups

# Age Group : Adult
vax_18_pvax = age['partial_18_29'] + age['partial_30_39'] + age['partial_40_49'] + age['partial_50_59']
print('Total Adult partially vaccinated: ',vax_18_pvax)

vax_18_fvax = age['full_18_29'] + age['full_30_39'] + age['full_40_49'] + age['full_50_59']
print('Total Adult fully vaccinated: ',vax_18_fvax)

vax_18_boost = age['booster_18_29'] + age['booster_30_39'] + age['booster_40_49'] + age['booster_50_59']
print('Total Adult received Booster: ',vax_18_boost)

# Age Group : +60 aged
vax_60_pvax = age['partial_60_69'] + age['partial_70_79'] + age['partial_80'] 
print('Total +60 aged partially vaccinated: ',vax_60_pvax)

vax_60_fvax = age['full_60_69'] + age['full_70_79'] + age['full_80']
print('Total +60 aged fully vaccinated: ',vax_60_fvax)

vax_60_boost = age['booster_60_69'] + age['booster_70_79'] + age['booster_80'] 
print('Total +60 aged received Booster: ',vax_60_boost)

# Alternative way of adding: using index
# vax_60_boost = sum(age[5:8])
```

```python
# Store values in a variable.
vax_12 = [age['partial_12_17'], age['full_12_17'], age['booster_12_17']]
vax_18 = [vax_18_pvax,vax_18_fvax, vax_18_boost]
vax_60 = [vax_60_pvax,vax_60_fvax, vax_60_boost]


# Define function for percentage calculation.
def cal(x , pop):
    perc = []
    for i in x:
        perc1 = i/pop*100
        perc.append(perc1)
    return perc
        
# Calculate % population vaccinated for each age group and store in a list as new variable.
perc_12 = cal(vax_12, ms_pop_12)
perc_18 = cal(vax_18, ms_pop_18)
perc_60 = cal(vax_60, ms_pop_60)

# Store lists in a new list.
data = [perc_12, perc_18, perc_60]

# Create a pd.DataFrame.
df = pd.DataFrame(data, columns = ['perc_pvax', 'perc_fvax','perc_boost'])
df.set_axis(['Adolescent', 'Adult', '60+ Aged'], axis=0, inplace=True)
df
```

```python
# Plot graph to visualise.

# plot a grouped bar chart.
ax = df[['perc_pvax','perc_fvax','perc_boost']].plot(kind = 'bar')
# change figure size with dimensions (width, height)
fig = ax.get_figure()
fig.set_size_inches(8, 7)

# add title, x and y-axes labels
plt.title('Vaccination Based on Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('% of Population')

# add legend by specifying the position 
plt.legend(['Partial','Fully','Booster'], bbox_to_anchor=(1, 1))


plt.show()
```

```python
# Using dataset 'total_vax'
list(total_vax)
```

```python
# Dataset from Github showed data only for 5 vaccine brands: pfizer, sinovac, AZ, sinopharm, cansino

# Calculate total vaccines administrated for each brand
pfizer = total_vax['pfizer1'] + total_vax['pfizer2'] + total_vax['pfizer3']
total_pfizer = sum(pfizer)
print('Total pfizer adminitrated: ',total_pfizer)

sinovac = total_vax['sinovac1'] + total_vax['sinovac2'] + total_vax['sinovac3']
total_sinovac = sum(sinovac)
print('Total sinovac adminitrated: ',total_sinovac)

az = total_vax['astra1'] + total_vax['astra2'] + total_vax['astra3']
total_az = sum(az)
print('Total pfizer adminitrated: ',total_az)

sinopharm = total_vax['sinopharm1'] + total_vax['sinopharm2'] + total_vax['sinopharm3']
total_sinopharm = sum(sinopharm)
print('Total sinopharm adminitrated: ',total_sinopharm)

cansino = total_vax['cansino'] + total_vax['cansino']
total_cansino = sum(cansino)
print('Total cansino adminitrated: ',total_cansino)

# Calculate total vaccines administrated for all brands
total_vax_ad = total_pfizer + total_sinovac + total_az + total_sinopharm + total_cansino
print('Total vaccine adminitrated: ',total_vax_ad)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining variables to plot.
vax_names = np.array(['Pfizer', 'Sinovac', 'AstraZeneca', 'Sinopharm', 'Cansino'])
vax_ad = np.array([total_pfizer,total_sinovac,total_az,total_sinopharm,total_cansino])

# Convert to percentage.
perc_tvax = []
for i in vax_ad:
    perc = i/total_vax_ad*100
    perc_tvax.append(perc)

# Plot pie chart to visualize.
plt.figure(figsize =(10, 7))
myexplode = [0.05,0,0,0.8,0]
mycolor = ['lightblue', 'orange', 'lightgreen', 'yellow', 'pink']
plt.pie(perc_tvax, labels = vax_names, startangle = 45, 
        shadow = True, autopct='%1.1f%%', explode = myexplode, colors = mycolor)
plt.title('Brands of Vaccine Administrated')

plt.tight_layout()
plt.show()
```

**New Approach**

```python
# Creating a new dataframe with selected attributes (cumulative data)
selected_columns = total_vax[['cumul_partial','cumul_full','cumul_booster']]
daily_vax = selected_columns.copy()
daily_vax
```

```python
# Calculate population % of each vaccination status

daily_vax['perc_pvax'] = daily_vax.cumul_partial/ms_total_pop*100
daily_vax['perc_fvax'] = daily_vax.cumul_full/ms_total_pop*100
daily_vax['perc_booster'] = daily_vax.cumul_booster/ms_total_pop*100
daily_vax
```

```python
# daily_icu dataframe starting date is 2020-03-24
# daily_vax dataframe staring date is 2021-02-24
# To plot two dataframes, standardize starting date by dropping data in daily_icu

# Identify which row for date 2021-02-24
# Convert rownames to a new column
daily_icu1 = daily_icu.rename_axis("date").reset_index()
# Identify the index number for the date 2021-02-24
a = daily_icu1.loc[daily_icu1['date'] == '2021-02-24']
print('The index number for the date 2021-02-24:\n' ,a)

# Drop the rows from index 0 - 336 in daily_icu dataframe
new_daily_icu = daily_icu.drop(daily_icu.index[0:337],0)
new_daily_icu
```

```python
# Plot daily ICU admission over time
fig, ax = plt.subplots(1,1,figsize=(10,6))
plt.plot(new_daily_icu, 'grey', label='ICU admission')
plt.legend(loc=2)
plt.xlabel('Date')
plt.ylabel('Number of ICU Admission')
plt.gcf().autofmt_xdate()

# Plot second graph on % population vaccinated over time in the previous graph 
ax = ax.twinx()
plt.plot(daily_vax['perc_pvax'], '--y', label = 'Partial')
plt.plot(daily_vax['perc_fvax'], 'g', label = 'Fully')
plt.plot(daily_vax['perc_booster'], '--b', label = 'Booster')
plt.ylim(0,100)
plt.legend(loc=1)
plt.ylabel('% Population Vaccinated')

plt.title('Comparison between ICU admission and Vaccination Status')

plt.show()
```
