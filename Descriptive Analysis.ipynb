```python
# Load total population in Malaysia dataset
ms_pop = pd.read_csv(folder + 'static/population.csv')
ms_pop
```

```python
# Store the Total Country Population in a variable
ms_total_pop = ms_pop['pop'].iloc[0]
ms_total_pop
```

```python
state_pop = ms_pop.drop([0], axis = 0)
state_pop.sort_values(by = ['state'], inplace = True, ascending = True)
state_pop
```

### Total population fully, partially vaccinated and received boosters

```python
# Load malaysia vaccination dataset
# This dataset includes all age groups, nationalities, ethinicities 
total_vax = pd.read_csv(folder + 'vaccination/vax_malaysia.csv')
total_vax
```

```python
list(total_vax)
```

```python
# Calculate vaccinated rate. 
total_pvax = total_vax['cumul_partial'].iloc[-1]
perc_pvax = total_pvax/ms_total_pop*100
print(perc_pvax, 'of population are partially vaccinated.')
total_fvax = total_vax['cumul_full'].iloc[-1]
perc_fvax = total_fvax/ms_total_pop*100
print(perc_fvax, 'of population are fully vaccinated.')
total_boost = total_vax['cumul_booster'].iloc[-1]
perc_boost = total_boost/ms_total_pop*100
print(perc_boost, 'of population received booster.')
```

```python
# Plot a graph to visualize
plt.figure(figsize=(10, 8))
group = ['Partial','Fully','Booster']
percentage = [perc_pvax, perc_fvax , perc_boost]
plt.bar(group, percentage, width = 0.5)
plt.ylim([0,100])
plt.axhline(y = 80, color = 'r')
def value_label(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha="center")
value_label(group,percentage)

plt.show()
```

### Total population fully, partially vaccinated and received boosters based on states

```python
# Load vaccination among age groups dataset
vax_state = pd.read_csv(folder + 'vaccination/vax_state.csv')
vax_state
```

```python
list(vax_state)
```

```python
state_pvax = vax_state.groupby('state')['daily_partial'].sum()
state_pvax = pd.DataFrame(state_pvax)

state_fvax = vax_state.groupby('state')['daily_full'].sum()
state_fvax = pd.DataFrame(state_fvax)

state_boost = vax_state.groupby('state')['daily_booster'].sum()
state_boost = pd.DataFrame(state_boost)

new_vax_state = pd.concat([state_pvax,state_fvax, state_boost], axis =1)

new_vax_state['state_population'] = state_pop['pop'].values
new_vax_state
```

```python
new_vax_state.index.name = 'state'
new_vax_state.reset_index(inplace=True)
new_vax_state
```

```python
# Calculate vaccinated rate. 
new_vax_state['perc_pvax'] = new_vax_state.daily_partial/new_vax_state.state_population*100
new_vax_state['perc_fvax'] = new_vax_state.daily_full/new_vax_state.state_population*100
new_vax_state['perc_boost'] = new_vax_state.daily_booster/new_vax_state.state_population*100
new_vax_state
```

```python
# Plot graph to visualise 

new_vax_state[['perc_pvax','perc_fvax','perc_boost']].plot(kind = 'bar')
plt.axhline(y = 100, color = 'r')
plt.title('Vaccination Based on States')
plt.xlabel('State')
plt.ylabel('% of Population')
plt.legend(['Partial','Fully','Booster'])
plt.show()
```
