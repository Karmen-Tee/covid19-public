*Plotting graph to visualise

```python
# Import library 
import matplotlib.pyplot as plt

# Plot graph
unvax_cases = daily_cases.cases_unvax
pvax_cases = daily_cases.cases_pvax
fvax_cases = daily_cases.cases_fvax
boost_cases = daily_cases.cases_boost
plt.figure(figsize=(10, 8))
plt.plot(unvax_cases)
plt.plot(pvax_cases)
plt.plot(fvax_cases)
plt.plot(boost_cases)
plt.title('Daily New Cases Among Unvaccinated, Partial, Fully Vaccinated and Booster Population')
plt.xlabel('Date')
plt.ylabel('Number of New Cases')
plt.legend(['Unvaccinated','Partial Vaccinated','Fully Vaccinated','Booster'])

# Making x-axis values (date) clearer
plt.gcf().autofmt_xdate()

plt.show()
```
